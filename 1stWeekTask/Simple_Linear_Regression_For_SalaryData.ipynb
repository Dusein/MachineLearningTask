Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Assign path dataset

```python
dataset = pd.read_csv('/content/sample_data/Salary_Data.csv')
```

Scatter Plot

```python
dataset.plot(kind='scatter',x='Salary',y='YearsExperience')
plt.show()
```

```python
# Membagi jadi training data dan test data
X_train, X_test, Y_train, Y_test = train_test_split(dataset.Salary, dataset.YearsExperience)
```

```python
# Visualisasi
plt.scatter(X_train,Y_train,label='Training Data',color='r',alpha=.7)
plt.scatter(X_test,Y_test,label='Testing Data',color='g',alpha=.7)
plt.legend()
plt.title('Test Train Split')
plt.show()
```

```python
# Linear model
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1),Y_train.values)
```

```python
# Prediksi tes data
prediction = LR.predict(X_test.values.reshape(-1,1))

# Visualisasi
plt.plot(X_test,prediction,label='Linear Regression',color='b')
plt.scatter(X_test,Y_test,label='Actual Test Data',color='g',alpha=.7)
plt.legend()
plt.show()
```
