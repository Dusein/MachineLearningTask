```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import pandas as pd

# Assign path dataset
file_path = '/content/sample_data/Position_Salaries.csv'
data = pd.read_csv(file_path)

# Assign independent variable (Level) and dependent variable (Salary)
X = data[['Level']].values
y = data['Salary'].values

# Buat fitur dari Polynominal (dibuat sampai pangkat 4)
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

# Membuat model polynomial regression
poly_reg_model = LinearRegression()
poly_reg_model.fit(X_poly, y)

# Prediksi
y_pred = poly_reg_model.predict(X_poly)

# Visualisasi
plt.scatter(X, y, color='red', label='Actual data')
plt.plot(X, y_pred, color='blue', label='Polynomial regression')
plt.title('Polynomial Regression: Level vs Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
